```python
import pandas as pd
import joblib
from tkinter import filedialog

# Use filedialog to select the data file
input_file_path = filedialog.askopenfilename(title="Select the cleaned data file", filetypes=[("Excel Files", "*.xlsx")])

# Load model
model = joblib.load("fake_news_classifier.pkl")

# Load new unlabeled dataset
data = pd.read_excel(input_file_path)

texts = data["text"]  # update with correct column name

# Convert to string, replace missing values
texts = texts.fillna("").astype(str)

# Predict labels
data["prediction"] = model.predict(texts)

# Use filedialog to select where to save the cleaned data
output_file_path = filedialog.asksaveasfilename(title="Save the cleaned data", defaultextension=".xlsx", filetypes=[("Excel Files", "*.xlsx")])

# Save the processed data to the selected location
data.to_excel(output_file_path, index=False)

print("Data processed and saved")
```
